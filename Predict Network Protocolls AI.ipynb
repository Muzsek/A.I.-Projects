```python
import pandas as pd

df = pd.read_csv("/root/.cache/kagglehub/datasets/jsrojas/labeled-network-traffic-flows-114-applications/versions/1/Unicauca-dataset-April-June-2019-Network-flows.csv")

print("Sorok, oszlopok száma: ", df.shape)
print("\nOszlopok:")
print(df.columns.tolist())

df.head()
```

```python
unique_application_protocolls = df['application_protocol'].unique()

for protocol in unique_application_protocolls:
    print(protocol)

df = df[df['application_protocol'] != 'Unknown']

print(df.shape[0])
```

```python
X = df[['avg_ps','f_pktTotalCount','b_pktTotalCount','min_piat','max_piat','avg_piat']]
y = df['application_protocol']

print(f"X shape {X.shape[::]}\ny shape: {y.shape[::]}")

print(X.iloc[:3])
print(y.iloc[:3])
```

```python
print(X.isna().sum())
print(y.isna().sum())
```

```python
counts = y.value_counts()
print(counts)
valid_classes = counts[counts >= 2]

X = X[y.isin(valid_classes.index)]
y = y[y.isin(valid_classes.index)]

print(X.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42,
    stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)

print(classification_report(y_test, y_pred_knn, zero_division=0))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100, random_state = 42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

print(classification_report(y_test, y_pred_rf, zero_division=0))
```

```python
import lightgbm as lgb
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

train_data = lgb.Dataset(X_train, label = y_train_enc)
test_data = lgb.Dataset(X_test, label = y_test_enc, reference=train_data)

params = {
    'objective': 'multiclass',
    'num_class': len(y_train.unique()),
    'metric': 'multi_logloss',
    'verbosity': -1
}

lgb_model = lgb.train(
    params,
    train_data,
    valid_sets=[test_data],
    valid_names=['test'],
    callbacks=[lgb.early_stopping(stopping_rounds=10)]
)

y_pred_lgb = lgb_model.predict(X_test)
y_pred_classes = le.inverse_transform(y_pred_lgb.argmax(axis=1))

print(classification_report(y_test, y_pred_classes, zero_division=0))
```
